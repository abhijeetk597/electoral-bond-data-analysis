```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Importing subscriber data
sub = pd.read_csv(r"data/subscriber_details.csv")

# Import party data
party = pd.read_csv(r"data/party_details.csv")
```

```python
sub.head()
```

```python
sub.columns
```

```python
# renaming columns as required
sub.rename(columns={'Sr No.': 'sr',
                    'Reference No  (URN)': 'urn',
                    'Journal Date': 'j_date',
                    'Date of\nPurchase': 'p_date',
                    'Date of Expiry': 'e_date',
                    'Name of the Purchaser': 'purchaser',
                    'Prefix': 'prefix',
                    'Bond\nNumber': 'bond_number',
                    'Denominations': 'denominations',
                    'Issue Branch Code': 'i_branch_code',
                    'Status': 'status'}, inplace=True)
```

```python
sub.columns
```

```python
# Dropping columns sr and issue teller
sub.drop(columns=["sr", "Issue Teller"], inplace=True)
```

```python
# Change date format from object to datetime
sub["j_date"] = pd.to_datetime(sub["j_date"])
sub["p_date"] = pd.to_datetime(sub["p_date"])
sub["e_date"] = pd.to_datetime(sub["e_date"])
```

```python
plt.hist(sub.p_date)
```

```python
# create new column b_id by combining prefix and bond number
sub["b_id"] = sub.apply(lambda row: row["prefix"] + str(row["bond_number"]), axis=1)
```

```python
sub.sample(3)
```

```python
sub.drop(columns=["prefix", "bond_number"], inplace=True)
```

```python
sub.status.value_counts()
```

```python
sub.denominations = sub.denominations.str.replace(",", "").astype("int")
sub.denominations.unique()
```

```python
# changing denomination unit to cr
sub.denominations = sub.denominations/10000000
sub.denominations.unique()
```

```python
party.head()
```

```python
# Branches where bonds were redeemed
party["Pay Branch\nCode"].unique()
```

```python
party.drop(columns=["Sr No.", "Account no. of\nPolitical Party", "Pay Branch\nCode", "Pay Teller"], inplace=True)
party.sample(3)
```

```python
# renaming columns as required
party.rename(columns={'Date of\nEncashment': 'encash_date',
                    'Name of the Political Party': 'party',
                    'Prefix': 'prefix',
                    'Bond\nNumber': 'bond_number',
                    'Denominations': 'denominations'}, inplace=True)

# Change date format from object to datetime
party["encash_date"] = pd.to_datetime(party["encash_date"])
```

```python
party.denominations = party.denominations.str.replace(",", "").astype("int")
party.denominations = party.denominations/10000000
party.denominations.unique()
```

```python
# create new column b_id by combining prefix and bond number
party["b_id"] = party.apply(lambda row: row["prefix"] + str(row["bond_number"]), axis=1)
```

```python
party.sample(5)
```

```python
party.rename(columns={'denominations': 'encash_denominations'}, inplace=True)
```

```python
df = pd.merge(sub, party[["party", "encash_date", "encash_denominations"]], left_on=sub.b_id, right_on=party.b_id)
df.sample(5)
```

```python
party_total = party.groupby(["party"])["encash_denominations"].sum()
party_total
```

```python
party["encash_year"] = party["encash_date"].dt.year
party.head(3)
```

```python
party_year = party.pivot_table(index="party", columns="encash_year", values="encash_denominations", aggfunc="sum")
party_year
```

```python
party_year_total = pd.merge(party_year, party_total, on=party_year.index).sort_values(by="encash_denominations", ascending=False).set_index("key_0")
party_year_total
```

```python
# Top 5 parties
z = party_year_total.iloc[:5,:-1].T
z
```

```python
# Top 5 parties year wise
plt.plot(z, marker="D")
plt.legend(z.columns)
```

```python
plt.barh(y=party_total.index, width=party_total)
# only top 5 are significant
```
